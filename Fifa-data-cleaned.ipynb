```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data cleaning WorldCups.csv


url = "D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/IH_Project-4/WorldCups.csv"
df = pd.read_csv(url)

df = df.dropna(how="all")

df = df.drop_duplicates()

text_cols = df.select_dtypes(include="object").columns
df[text_cols] = df[text_cols].apply(lambda col: col.str.strip())

df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)

df.to_csv(r"D:\04_IronHack_DA\IRONHACK_LABS_PROJECTS\IH_Project-4_backup\WorldCups_cleaned.csv")
```

```python
# data cleaning WorldCupPlayers.csv


url2 = "D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/IH_Project-4/WorldCupPlayers.csv"

df2 = pd.read_csv(url2)

df2 = df2.dropna(how="all")

df2 = df2.drop_duplicates()

text_cols = df2.select_dtypes(include="object").columns
df2[text_cols] = df2[text_cols].apply(lambda col: col.str.strip())

df2.columns = df2.columns.str.lower().str.strip().str.replace(" ", "_").str.replace("-", "_")

df2 = df2.drop(columns=[
    "line_up",
    "shirt_number",
    "position",
    "event"
], errors="ignore")


df2.insert(0, "player_id", range(1, len(df2) + 1))

df2.to_csv(r"D:\04_IronHack_DA\IRONHACK_LABS_PROJECTS\IH_Project-4_backup\WorldCupPlayers_cleaned.csv")
```

```python
# data cleaning WorldCupMatches.csv


url3 = "D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/IH_Project-4/WorldCupMatches.csv"
df3 = pd.read_csv(url3)

df3 = df3.dropna(how="all")

text_cols = df3.select_dtypes(include="object").columns
df3[text_cols] = df3[text_cols].apply(lambda col: col.str.strip())

df3 = df3.drop(columns=[
    "Win conditions",
    "Referee",
    "Assistant 1",
    "Assistant 2"
])

df3["Datetime"] = df3["Datetime"].str.replace(" - ", " ", regex=False)
df3["Datetime"] = pd.to_datetime(df3["Datetime"], errors="coerce")

df3["Date"] = df3["Datetime"].dt.date
df3["Time"] = df3["Datetime"].dt.time
df3 = df3.drop(columns=["Datetime",])

df3.to_csv(r"D:\04_IronHack_DA\IRONHACK_LABS_PROJECTS\IH_Project-4_backup\WorldCupMatches_cleaned.csv")
```
