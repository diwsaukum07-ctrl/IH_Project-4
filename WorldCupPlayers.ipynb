```python
import pandas as pd
```

```python
df = pd.read_csv("WorldCupPlayers.csv")
```

```python
df.head()
```

```python
df = df.dropna(how="all")
```

```python
df = df.drop_duplicates()
```

```python
text_cols = df.select_dtypes(include="object").columns
df[text_cols] = df[text_cols].apply(lambda col: col.str.strip())
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)
```

```python
df.columns
```

```python
df = df.drop(columns=[
    "Line-up",
    "Shirt Number",
    "Position",
    "Event"
])
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_").str.replace("-", "_")

df = df.drop(columns=[
    "line_up",
    "shirt_number",
    "position",
    "event"
], errors="ignore")
```

```python
df.head()
```

```python
df.insert(0, "player_id", range(1, len(df) + 1))
```

```python
df.head()
```

```python
df.to_csv("WorldCupPlayers_cleaned.csv", index=False)
```
