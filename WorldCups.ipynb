```python
import pandas as pd
```

```python
df = pd.read_csv("WorldCups.csv")
```

```python
df.head()
```

```python
df = df.dropna(how="all")
```

```python
df = df.drop_duplicates()
```

```python
text_cols = df.select_dtypes(include="object").columns
df[text_cols] = df[text_cols].apply(lambda col: col.str.strip())
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)
```
