```python
import pandas as pd
```

```python
df = pd.read_csv("WorldCupMatches.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.dropna(how="all")
```

```python
df.shape
```

```python
text_cols = df.select_dtypes(include="object").columns
df[text_cols] = df[text_cols].apply(lambda col: col.str.strip())
```

```python
df.columns
```

```python
df = df.drop(columns=[
    "Win conditions",
    "Referee",
    "Assistant 1",
    "Assistant 2"
])
```

```python
df.columns
```

```python
df["Datetime"] = df["Datetime"].str.replace(" - ", " ", regex=False)
df["Datetime"] = pd.to_datetime(df["Datetime"], errors="coerce")
```

```python
df["Date"] = df["Datetime"].dt.date
df["Time"] = df["Datetime"].dt.time
```

```python
df.head()
```

```python
df = df.drop(columns=[
    "Datetime",
])
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.to_csv("WorldCupMatches_cleaned.csv", index=False)
```
